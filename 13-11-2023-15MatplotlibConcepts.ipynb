```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xp=np.array([0,6])
yp=np.array([0,50])
plt.plot(xp,yp)
plt.show()
```

```python
xp=np.array([0,6])
yp=np.array([0,50])
plt.plot(xp,yp,'o')
plt.show()
```

```python
xp=np.array([0,1,2,3,4])
yp=np.array([3,6,8,9,10])
plt.plot(xp,yp)
plt.show()
```

```python
y=np.array([3,8,1,10])
plt.plot(y,marker='H')
plt.show()
```

```python
y=np.array([3,8,1,10])
plt.plot(y,'o:r')
plt.show()
```

```python
x=[10,20,30,40]
y=[50,60,70,80]
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4])
plt.plot([10,20,30,40])
plt.plot([1,2,3,4,5],[10,20,30,40,50])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x=[10,20,30,40]
plt.plot(x,x,'ro')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xlim(0,30)
plt.ylim(0,30)
plt.plot(x)
plt.grid(True)
plt.show()
```

```python
t1=np.arange(5)
plt.plot(t1,'r--',t1**4,'bs',t1**3,'g^',t1**5,'ro')
plt.show()
```

```python
x=[1,2,3,4]
plt.plot(x,[elem for elem in x],label='linear')
plt.plot(x,[elem**2 for elem in x],label='square')
plt.grid(True)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.savefig('myimg.png')
plt.show()
```

```python
#Histogram
data=np.random.randn(8,8)
plt.hist(data)
plt.show()
```

```python
x=[10,20,30]
plt.hist(x)
plt.show()
```

```python
my_dict={'A':30,'B':65,'C':50,'D':80}
for i,key in my_dict.items():
    print(i,key)
    plt.bar(i,key)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#Creating a Pie Chart
x=[6000,20000,3000,70000,30000]
my_label=['s1','s2','s3','s4','s5']
plt.pie(x,labels=my_label)
plt.show()
```

```python
#Scatter Plot
x=np.random.rand(100)
y=np.random.rand(100)
plt.scatter(x,y)
plt.show()
```
